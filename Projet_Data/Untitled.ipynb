```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
myData = pd.read_csv('C:\\Users\\kanga\\Desktop\\Projet_Data\\Ottawa_Data.csv', index_col = 'Date/Time', parse_dates = True)   #read_csv == fonction qui permet de lire un fichier
```

```python
# Afficher les premières lignes pour comprendre la structure des données
myData.head(5)
```

```python
# Vérifier les informations sur les données
myData.info()
```

```python
# Vérifier les statistiques descriptives
myData.describe()
```

```python
# Supprimer les lignes avec des valeurs manquantes
myData.isna().sum()
```

```python
for col in myData.select_dtypes(include = ['float', 'int']):
    myData[col] = myData[col].fillna(myData[col].mean())

#remplisage des valeur NAN pour chaque chauque collones specifique 
myData['Max Temp (°C)'] = myData['Max Temp (°C)'].fillna(myData['Max Temp (°C)'].mean())
myData['Min Temp (°C)'] = myData['Min Temp (°C)'].fillna(myData['Min Temp (°C)'].mean())
myData['Total Rain (mm)'] = myData['Total Rain (mm)'].fillna(myData['Total Rain (mm)'].mean())
myData['Total Snow (cm)'] = myData['Total Snow (cm)'].fillna(myData['Total Snow (cm)'].mean())
myData['Total Precip (mm)'] = myData['Total Precip (mm)'].fillna(myData['Total Precip (mm)'].mean())
myData['Snow on Grnd (cm)'] = myData['Snow on Grnd (cm)'].fillna(myData['Snow on Grnd (cm)'].mean())
myData['Dir of Max Gust (10s deg)'] = myData['Dir of Max Gust (10s deg)'].fillna(myData['Dir of Max Gust (10s deg)'].mean())
myData['Spd of Max Gust (km/h)'] = myData['Spd of Max Gust (km/h)'].fillna(myData['Spd of Max Gust (km/h)'].mean())
```

```python
# Supprimer les lignes avec des valeurs manquantes
myData.dropna(axis=1, thresh=len(myData)*0.5, inplace=True)
```

```python
myData.head(5)
```

```python
# Moyenne mensuelle des températures maximales et minimales
monthly_temps = myData.groupby(myData.index.month)[['Max Temp (°C)', 'Min Temp (°C)']].mean()
monthly_temps.plot()
```

```python
# Total des précipitations par mois
precip_monthly = myData.groupby(myData.index.month)['Total Precip (mm)'].sum()
precip_monthly.plot(kind='bar')
```

```python
# Jours avec températures extrêmes
extreme_high = myData[myData['Max Temp (°C)'] > 30]
extreme_low = myData[myData['Min Temp (°C)'] < -10]
```

```python
# Trouver les jours avec des précipitations élevées
high_precipitation_days = myData[myData['Total Precip (mm)'] > 50]
```

```python
# Tracer des graphiques pour visualiser les données
plt.figure(figsize=(10, 5))
plt.plot(myData['Mean Temp (°C)'], label='Température Moyenne')
plt.title('Température Moyenne Quotidienne')
plt.xlabel('Date')
plt.ylabel('Température (°C)')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(myData['Total Precip (mm)'], color = 'Orange',label='Précipitations Totales')
plt.title('Précipitations Quotidiennes')
plt.xlabel('Date')
plt.ylabel('Précipitations (mm)')
plt.legend()
plt.show()
```

```python
# Calculer la matrice de corrélation
correlation_matrix = myData[['Max Temp (°C)', 'Total Precip (mm)']].corr()
print(correlation_matrix)
```
